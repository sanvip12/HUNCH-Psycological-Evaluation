Logistic Regression CLassification

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import time
import joblib

data = pd.read_csv(r"C:\Users\Anjali\Downloads\datasets_female.csv", delimiter='\t')
data.columns = data.columns.str.strip()


X = data[['Positive', 'Neutral', 'Negative', 'Pitch', 'Pitch Variance', 'Amplitude']] 
y = data['Emotion'] 


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000, C=100, class_weight='balanced')


start_time = time.time()
model.fit(X_scaled, y)
end_time = time.time()
training_time = end_time - start_time   

y_pred = model.predict(X_scaled)


accuracy = accuracy_score(y, y_pred)


print(f"Training Time: {training_time:.2f} seconds")
print("\nConfusion Matrix:")
print(confusion_matrix(y, y_pred))
print("\nClassification Report:")
print(classification_report(y, y_pred))
print(f"Overall Accuracy: {accuracy:.4f}")

joblib.dump(model, 'EmotionAIModel.joblib')
joblib.dump(scaler, 'scaler.joblib')

print("\nModel and scaler saved successfully!")
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
import time
import joblib

data = pd.read_csv(r"C:\Users\Anjali\Downloads\datasets_female.csv", delimiter='\t')
data.columns = data.columns.str.strip()


X = data[['Positive', 'Neutral', 'Negative', 'Pitch', 'Pitch Variance', 'Amplitude']] 
y = data['Emotion']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


start_time = time.time()

rf_model = RandomForestClassifier(n_estimators=100, max_depth=3, random_state=42)
rf_model.fit(X_scaled, y)

end_time = time.time()
training_time = end_time - start_time

y_pred = rf_model.predict(X_scaled)


accuracy = accuracy_score(y, y_pred)

print(f"Training Time: {training_time:.2f} seconds")
print("\nConfusion Matrix:")
print(confusion_matrix(y, y_pred))
print("\nClassification Report:")
print(classification_report(y, y_pred))
print(f"Overall Accuracy: {accuracy:.4f}")


joblib.dump(rf_model, 'RandomForestEmotionModel.joblib')
joblib.dump(scaler, 'scaler.joblib')

print("\nModel and scaler saved successfully!")
```
